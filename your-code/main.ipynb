#### Question 1: Find the questions and answers of last month.

```python
# your code here
SITE = StackAPI('stackoverflow')
comments = SITE.fetch('comments')
# Does not work when I add 'fromdate=datetime(2019,7,10), todate=datetime(2019,7,11)' after comments

data = pd.DataFrame(comments)
print(data)
```

```python
def flatten (data, col_list):
    for column in col_list:
        flattened = pd.DataFrame(dict(data[column])).transpose()
        columns = [str(col) for col in flattened.columns]
        flattened.columns = [column + '_' + colname for colname in columns]
        data = pd.concat([data, flattened], axis=1)
        data = data.drop(column, axis=1)
    return data

nested_columns = ['items']

flat = flatten(data, nested_columns)
flat
```

```python
nested_column = ['items_owner']
flatter = flatten(flat, nested_column)
flatter
```

```python
flattest = flatten(flatter, ["items_reply_to_user"])
flattest
```

```python
questions_answers = flattest[["items_comment_id", "items_reply_to_user_user_id"]]
questions_answers
# The comment and reply IDs below
```
